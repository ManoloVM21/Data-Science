```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix

# Reload clean copy of the data (or use your existing 'campaign' if it is untouched)
campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')

# Drop obvious missing rows (this dataset has very few)
campaign = campaign.dropna().copy()

campaign.head()
campaign['y'].value_counts(normalize=True)
```

```
# Target: did the client subscribe to a term deposit?
y = campaign['y'].map({'no': 0, 'yes': 1})   # convert to 0/1

# Drop target to get features
X = campaign.drop(columns=['y'])

# Identify numeric vs categorical columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

numeric_cols, categorical_cols
```

```
# One‑hot encode categoricals, leave numerics as-is
preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', 'passthrough', numeric_cols)
    ]
)

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'  # very important: target is imbalanced
)

model = Pipeline(steps=[
    ('preprocess', preprocess),
    ('rf', rf)
])
```

```
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

model.fit(X_train, y_train)
```

```
y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=3))
```

```
# Get feature names after one‑hot encoding
ohe = model.named_steps['preprocess'].named_transformers_['cat']
ohe_features = ohe.get_feature_names_out(categorical_cols)
all_features = list(ohe_features) + numeric_cols

importances = model.named_steps['rf'].feature_importances_

feat_imp = pd.DataFrame({
    'feature': all_features,
    'importance': importances
}).sort_values('importance', ascending=False)

feat_imp.head(20)
```

```
## Load the MINI Holdout Data

holdout = pd.read_csv("https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank_holdout_test_mini.csv")
holdout.head()
```

```
## Run Mini Holdout through our model

holdout_preds = model.predict(holdout)
holdout_probs = model.predict_proba(holdout)[:, 1]  # probability of class 1 ("yes")

results = holdout.copy()
results["predicted_y"] = holdout_preds
results["predicted_prob_yes"] = holdout_probs
results.head()
```

```
# Just the predicted class (0/1)
preds_only = pd.DataFrame({"predictions": holdout_preds})
preds_only.to_csv("team4-module2-predictions.csv", index=False)
```
